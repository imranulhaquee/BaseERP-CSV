```python
#selRegion = request.POST.get('num4', 'GCC')
#selRegion = "GCC"

if( selRegion == "GCC"):
    tData = pd.read_csv("media/database/TemporaryGCC.csv" ,usecols=req_cols)
    aData = pd.read_csv("DataBase/1_Sales/"+cYear+" "+selRegion+" "+"Sales.csv" ,usecols=req_cols)
    Act_frames = (tData , aData)
    cData = pd.concat(Act_frames)
elif( selRegion == "Levant"):
    tData = pd.read_csv("media/database/TemporaryLEB.csv" ,usecols=req_cols)
    aData = pd.read_csv("DataBase/1_Sales/"+cYear+" "+selRegion+" "+"Sales.csv" ,usecols=req_cols)
    Act_frames = (tData , aData)
    cData = pd.concat(Act_frames)
elif( selRegion == "MCS"):
    tData = pd.read_csv("media/database/TemporaryMCS.csv" ,usecols=req_cols)
    aData = pd.read_csv("DataBase/1_Sales/"+cYear+" "+selRegion+" "+"Sales.csv" ,usecols=req_cols)
    Act_frames = (tData , aData)
    cData = pd.concat(Act_frames)
elif( selRegion == "MCS-2"):
    tData = pd.read_csv("media/database/TemporaryMCS-2.csv" ,usecols=req_cols)
    aData = pd.read_csv("DataBase/1_Sales/"+cYear+" MCS Sales-2.csv" ,usecols=req_cols)
    Act_frames = (tData , aData)
    cData = pd.concat(Act_frames)
else:
    pass
```
